```python
# 输入数据
INPUT_DATA = "blackmetal/data/shuju1.csv"
```

```python
raw_data = pd.read_csv(INPUT_DATA)
```

```python
raw_data.columns = ["date", "time", "open", "close", "high", "low", "volume", 
                     "p1", "p2", "p3", "p4", "p5", "p6", "p7", "p8", "p9", "p10",
                     "p11", "p12", "p13", "p14", "p15", "p16", "p17", "p18", "p19", "p20", 
                     "p21", "p22", "p23", "p24", "p25", "p26", "p27", "p28", "p29", "p30",
                     "p31", "p32", "p33", "p34", "p35", "p36", "p37"]
```

```python
labels = raw_data[["date", "time","high"]]
labels_tag = labels["high"]
```

```python
returns = 100 * labels_tag.pct_change().dropna()
%matplotlib inline 
```

```python
returns.plot(figsize=(20,6))
```

```python
from arch import arch_model
am = arch_model(returns) 
res = am.fit(update_freq=5)
```

```python
print(res.summary())
```

```python
fig = res.plot(annualize='D')
```

```python
am = arch_model(returns, p=1, o=1, q=1)
res = am.fit(update_freq=5, disp='off')
print(res.summary())
```

```python
am = arch_model(returns, p=1, o=1, q=1, power=1.0)
res = am.fit(update_freq=5)
print(res.summary())
```

```python
am = arch_model(returns, p=1, o=1, q=1, power=1.0, dist='StudentsT')
res = am.fit(update_freq=5)
print(res.summary())
```

```python
fixed_res = am.fix([0.0235, 0.01, 0.06, 0.0, 0.9382, 8.0])
print(fixed_res.summary())
```

```python
import pandas as pd
df = pd.concat([res.conditional_volatility,fixed_res.conditional_volatility],1)
df.columns = ['Estimated', 'Fixed']
df.plot()
```
